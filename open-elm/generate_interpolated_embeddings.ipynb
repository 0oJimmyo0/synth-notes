```python
import pickle
import random
import numpy as np

random.seed(168)  # set random seed for reproducibility
```

```python
# load the test dataset
with open(f'../data/pubmed_rct/processed_dataset/test_with_embeddings.pkl', 'rb') as f:
    test_dict = pickle.load(f)
test_abstracts = test_dict["abstracts"]
test_embeddings = test_dict["embeddings"]
```

```python
test_embeddings.shape
```

```python
# create a interpolated dataset
interpolated_abstract_i = []
interpolated_abstract_j = []
interpolated_embs = []

for i in range(len(test_abstracts)):
    
    emb_i = test_embeddings[i]

    # randomly select the other embedding as emb_j
    j = random.randint(0, len(test_abstracts) - 1)
    while j == i:  # ensure we don't select the same abstract
        j = random.randint(0, len(test_abstracts) - 1)
    
    emb_j = test_embeddings[j]
    interpolated_emb = (emb_i + emb_j) / 2
    
    interpolated_abstract_i.append(test_abstracts[i])
    interpolated_abstract_j.append(test_abstracts[j])
    interpolated_embs.append(interpolated_emb)

# save as pickle
interpolated_data = {
    "abstract_i": interpolated_abstract_i,
    "abstract_j": interpolated_abstract_j,
    "embeddings": np.array(interpolated_embs)
}
```

```python
interpolated_data["embeddings"].shape
```

```python
with open('../data/pubmed_rct/processed_dataset/interpolated_embeddings.pkl', 'wb') as f:
    pickle.dump(interpolated_data, f)
```
